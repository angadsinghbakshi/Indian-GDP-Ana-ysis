# Unemployment Rate Trends by City (2019-Present)

```python
cities  = df['City'].unique()

fig, ax = plt.subplots(5,3, figsize = (25,25), sharex = True, sharey = True)
ax = ax.flatten()

sns.set_palette('flare')
palette = sns.color_palette()
sixth = palette[5]

for i, city in enumerate(cities):
    city_data = df[(df['City'] == city)]
    sns.lineplot(x = city_data['Year'],
                 y = city_data['Unemployment Rate (%)'],
                 label=city, ax= ax[i], marker = 'o',linewidth = 3, color = sixth)
    ax[i].set_title(city, weight = 'bold')
```

## It can be observed that post covid most of the states saw a decline in unemployment rate and very few states have shown an increase in unemployment rate.
## Ahmedabad and Mumbai have the highest unemployment rate currently.

## GDP Change from 2019 to 2023

```python
gdp = df[(df['Year'] == 2019) | (df['Year'] == 2023)]
gdp['Year'] = gdp['Year'].astype(str)

fig, axes = plt.subplots(5,3, figsize =  (24,24), sharex = True, sharey = True)
axes = axes.flatten()

palette = sns.color_palette('pastel')
ff = palette[4:6]

for i,city in enumerate(cities):
    gdp_city = gdp[(gdp['City'] == city)]
    ax = sns.barplot(data = gdp_city, x = 'Year', y = 'GDP (in billion $)',
                     label = city, ax = axes[i], legend = False,
                     hue = 'Year', palette = ff)
    
    for container in ax.containers:
        ax.bar_label(container, fontsize = 12, weight = 'bold')
        
    axes[i].set_title(city, weight = 'bold')
```

## Only Vadodara and Lucknow have a drop in their GDPs.
## Vadodara had a signficant drop in GDP.
## Bhopal had the maximum increase in GDP over 5 years period.

```python
gdp2 = gdp[(gdp['City'] == 'Bhopal') | (gdp['City'] == 'Vadodara')]
```

```python
fig, axx = plt.subplots(1,2, figsize = (12,8), sharex = True, sharey = True)

for i,city in enumerate(gdp2['City'].unique()):
    bho_vad = gdp2[gdp2['City'] == city]
    ax = sns.barplot(data = bho_vad, x = 'Year', y = 'GDP (in billion $)',
                ax = axx[i], hue = 'Year', label = city, legend = False,
                    palette = ff)
    for container in ax.containers:
        ax.bar_label(container)
    axx[i].set_title(city, weight = 'bold')
```

## Youth Unemployment Rate in Tier 1 Cities

```python
selected_cities = ['Bengaluru', 'Delhi', 'Chennai', 'Hyderabad', 'Kolkata', 'Mumbai', 'Pune', 'Ahmedabad']

cities = df[(df['City'] == 'Bengaluru') | (df['City'] == 'Delhi') | (df['City'] == 'Chennai') | (df['City'] == 'Hyderabad') | (df['City'] == 'Kolkata') | (df['City'] == 'Mumbai') | (df['City'] == 'Pune') | (df['City'] == 'Ahmedabad')]
fig, ax = plt.subplots(4,2,  figsize = (25,25), sharex = True, sharey = True)
ax = ax.flatten()

sns.set_palette('pastel')
palette = sns.color_palette()
sixth = palette[3]

for i, city in enumerate(selected_cities):
    city_data = cities[(cities['City'] == city)]
    sns.lineplot(x = city_data['Year'],
                 y = city_data['Youth Unemployment Rate (%)'],
                 label=city, ax= ax[i], marker = 'o',linewidth = 3, color = sixth,
                legend = False)
    ax[i].set_title(city, weight = 'bold')
```


## It can be observed that the youth unemployment rate in Tier 1 cities like Bangalore, Chennai, Kolkata and Ahmedabad are showing an increase and the rest of the cities are showing a decrease in Youth Unemployment rate.

## Top Cities with Best numbers in Argiculture, Industrym Services and Technology each Year

```python
sectors = df[['Industry (%)' , 'Services (%)' , 'Technology (%)','Agriculture (%)']]
for year in df['Year'].unique():
    print(year,"\n")
    df_year = df[df['Year'] == year]
    for sector in sectors:
        city = df.loc[df_year[sector].idxmax(), 'City']
        pct = df_year[sector].max()
        print(f"{sector}: {city}, ({pct}%)")
    print("\n")
```

## It can be observed that each year, for every Sector we had a different City being on top which can be an indicator that cities are developing and competing with each other and this will help in the overall progress of the Country.

## Lowest cities in terms of unemployment Rate

```python
sectors = df[['Unemployment Rate (%)','Youth Unemployment Rate (%)']]
for year in df['Year'].unique():
    print(year,"\n")
    df_year = df[df['Year'] == year]
    for sector in sectors:
        city = df.loc[df_year[sector].idxmin(), 'City']
        pct = df_year[sector].max()
        print(f"{sector}: {city}, ({pct}%)\n")
    print("\n")
    
```
